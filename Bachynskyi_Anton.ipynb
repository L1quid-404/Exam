### Теоретичні відомості: Розгалуження
Алгоритми з розгалуженням передбачають виконання різних послідовностей дій залежно від виконання певних умов. В мові Python для цього використовуються оператори:
- `if`: основна умова.
- `elif` (else if): додаткові умови, якщо попередня не виконалась.
- `else`: виконується, якщо жодна з попередніх умов не є істинною.

### Завдання 1: Класифікація числа
Програма приймає ціле число та визначає, до якої категорії воно належить: додатне, від'ємне або нуль.

```python
x = 42

if x > 0:
    print("Число є додатним")
elif x < 0:
    print("Число є від'ємним")
else:
    print("Число дорівнює нулю")
```

### Теоретичні відомості: Винятки
Винятки (Exceptions) — це помилки, які виникають під час виконання вже скомпільованої програми. Для того, щоб програма не завершувалася аварійно, використовується конструкція `try...except`:
- `try`: блок коду, в якому може виникнути помилка.
- `except`: блок коду, який виконується лише у разі виникнення помилки певного типу.
- `finally`: блок, який виконується в будь-якому випадку (зазвичай для закриття ресурсів).

### Завдання 2: Безпечне ділення
Реалізувати програму, яка виконує ділення двох чисел. У випадку спроби ділення на нуль, програма має перехопити виняток `ZeroDivisionError` та вивести зрозуміле повідомлення замість системної помилки.

```python
dividend = 100
divisor = 0

try:
    result = dividend / divisor
    print(result)
except ZeroDivisionError:
    print("Помилка: Ділення на нуль заборонене правилами математики")
```

### Завдання 3: Перевірка введених даних
Програма перевіряє, чи є введене значення числом, і якщо так — перевіряє його на парність.

```python
user_input = "15"

try:
    number = int(user_input)
    if number % 2 == 0:
        print("Число парне")
    else:
        print("Число непарне")
except ValueError:
    print("Помилка: Введене значення не є коректним числом")
```
